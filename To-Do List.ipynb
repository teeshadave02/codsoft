# Task:3 To-Do List

```python
import tkinter as tk

def add_task():
    task = entry.get()
    if task:
        tasks.append(task)
        update_tasks()
        entry.delete(0, tk.END)

def delete_task():
    selected_task = task_listbox.curselection()
    if selected_task:
        task_index = int(selected_task[0])
        del tasks[task_index]
        update_tasks()

def update_task():
    selected_task = task_listbox.curselection()
    if selected_task:
        task_index = int(selected_task[0])
        new_task = entry.get()
        if new_task:
            tasks[task_index] = new_task
            update_tasks()
            entry.delete(0, tk.END)

def update_tasks():
    task_listbox.delete(0, tk.END)
    for task in tasks:
        task_listbox.insert(tk.END, task)

# Create the main window
root = tk.Tk()
root.title("To-Do List App")

tasks = []

# Create an entry for adding/updating tasks
entry = tk.Entry(root, width=30)
entry.pack(padx=10, pady=5)

# Create a frame for the buttons
button_frame = tk.Frame(root)
button_frame.pack(padx=10, pady=5)

# Create buttons for adding, updating, and deleting tasks
add_button = tk.Button(button_frame, text="Add Task", command=add_task)
add_button.grid(row=0, column=0, padx=5)

update_button = tk.Button(button_frame, text="Update Task", command=update_task)
update_button.grid(row=0, column=1, padx=5)

delete_button = tk.Button(button_frame, text="Delete Task", command=delete_task)
delete_button.grid(row=0, column=2, padx=5)

# Create a Listbox to display tasks
task_listbox = tk.Listbox(root, width=40)
task_listbox.pack(padx=10, pady=5)

# Load existing tasks if any
try:
    with open("tasks.txt", "r") as file:
        for line in file:
            tasks.append(line.strip())
    update_tasks()
except FileNotFoundError:
    pass

# Save tasks to a file when closing the app
def on_closing():
    with open("tasks.txt", "w") as file:
        for task in tasks:
            file.write(task + "\n")
    root.destroy()

root.protocol("WM_DELETE_WINDOW", on_closing)

root.mainloop()
```
